```r
library(tidyverse)
library(plotly)
library(circlize)
```

# Sepsis Chord Plot

```r
# Load in the summary data for the untargeted metabolomics into dataframe

sepsis_tissues_metab  <- read.csv('/Users/brooksleitner/Desktop/Python/sepsismetabolism/data/metabolomics/sepsis_tissue_metab_chord.csv')
```

```r
sepsis_chord_df <- sepsis_tissues_metab[, c('TissueA', 'TissueB', 'Total_Sig_Corrs')]
```

```r
chordDiagram(sepsis_chord_df, directional = 2)
```

```r

grid.col = c(eWAT = "forestgreen", Heart = "coral1", Kidney = "dodgerblue",
    Liver = "cadetblue1", Lung = "grey", Quad = "darkorange", Spleen = "darkorchid1")
chordDiagram(sepsis_chord_df, directional = 2, grid.col = grid.col, transparency = .8)
```

```r
# Load in the summary data for the untargeted metabolomics into dataframe
#This dataframe has the directionality

sepsis_tissues_metab_dir  <- read.csv('/Users/brooksleitner/Desktop/Python/sepsismetabolism/data/metabolomics/sepsis_tissue_metab_chord_directions.csv')
```

```r
sepsis_chord_df_dir <- sepsis_tissues_metab_dir[, c('TissueA', 'TissueB', 'Total_Sig_Corrs')]
```

```r
svg("/Users/brooksleitner/Desktop/Python/sepsismetabolism/figures/metabolomics/sepsis_tissue_chord.svg")

col = c('coral1', 'lightblue')
grid.col = c(eWAT = "forestgreen", Heart = "rosybrown1", Kidney = "dodgerblue",
    Liver = "cadetblue1", Lung = "khaki1", Quad = "coral1", Spleen = "mediumpurple1")
chordDiagram(sepsis_chord_df_dir, directional = 2, grid.col = grid.col, col=col, transparency = 0.5)

# Close the graphics device
dev.off() 
```
